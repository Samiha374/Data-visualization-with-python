```python
import pandas as pd
import matplotlib.pyplot as plt
url = "company_sales_data.csv"
df = pd.read_csv(url)
```

```python
df.dropna(inplace=True)
print(df.isnull().sum())
```

```python
#Exercise 1
plt.plot(df['month_number'], df['total_profit'])
plt.xlabel('Month Number')
plt.ylabel('Total Profit ($)')
plt.title('Total Profit of each Months')
plt.show()
```

```python
#Exercise 2
plt.plot(df['month_number'], df['total_profit'], linestyle='dotted', color='red', marker='o', linewidth='3', label='Total Profit')
plt.xlabel('Month Number')
plt.ylabel('Sold unnit number')
plt.title('Total Profit of each Months')
plt.show()
```

```python
#Exercise 3
plt.plot(df['month_number'], df['facecream'], label='Face Cream')
plt.plot(df['month_number'], df['facewash'], label='Facewash')
plt.plot(df['month_number'], df['toothpaste'], label='Toothpaste')
plt.plot(df['month_number'], df['bathingsoap'], label='Bathing Soap')
plt.plot(df['month_number'], df['shampoo'], label='Shampoo')
plt.plot(df['month_number'], df['moisturizer'], label='Moisturizer')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.legend()
plt.title('Product Sales Data')
plt.show()
```

```python
#Exercise 4
plt.scatter(df['month_number'], df['toothpaste'])
plt.grid(True, linestyle='-')
plt.xlabel('Month Number')
plt.ylabel('Toothpaste Sales')
plt.title('Toothpaste Sales Grid')
plt.show()
```

```python
#Exercise 5
plt.bar(df['month_number'], df['facecream'], label='Face Cream')
plt.bar(df['month_number'], df['facewash'], label='Facewash')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.legend()
plt.title('Face Cream and Facewash Sales')
plt.show()
```

```python
#Exercise 6
plt.bar(df['month_number'], df['bathingsoap'])
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Bathing Soap Sales')
plt.savefig('bathing_soap_sales.png') #
plt.show()
```

```python
#Exercise 7
profitList = df['total_profit'].tolist()
profit_range = [150000,175000,2000000,225000,250000,300000,350000]
plt.hist(profitList,profit_range,label='profit data')
plt.xlabel('Profit Range ($)')
plt.ylabel('Actual Profit')
plt.legend(loc='upper left')
plt.xticks(profit_range)
plt.title('profit data')
plt.show()
```

```python
#Exercise 8
product_columns = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
total_sale_data = df[product_columns].sum()
plt.pie(total_sale_data, labels=total_sale_data.index, autopct='%1.1f%%')
plt.title('Total Sale Data')
plt.show()
```

```python
#Exercise 9
fig, (ax1, ax2) = plt.subplots(2, 1)
ax1.bar(df['month_number'], df['bathingsoap'])
ax1.set_ylabel('Bathing Soap Sales')

ax2.bar(df['month_number'], df['facewash'])
ax2.set_ylabel('Facewash Sales')
ax2.set_xlabel('Month Number')
plt.show()
```

```python
#Exercise 10
plt.stackplot(df['month_number'], df['facecream'], df['facewash'], df['toothpaste'], df['bathingsoap'], df['shampoo'], df['moisturizer'], labels=['Face Cream', 'Facewash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer'])
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.legend(loc='upper left')
plt.title('Stack Plot for All Product Sales Data')
plt.show()
```
