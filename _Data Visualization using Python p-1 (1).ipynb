```python
import numpy as np  # useful for many scientific computing in Python
import pandas as pd
%matplotlib inline 
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df_sal=pd.read_csv("/Users/ykwalled/Downloads/company_sales_data.csv")
print('Data read into a pandas dataframe!')
```

```python
df_sal.head()
```

```python
df_sal[['total_profit']]
df_sal[['total_profit']].head(5)
```

```python
 #Read Total profit of all months and show it using a line plot
#Total profit data provided for each month. Generated line plot must include the following properties: –
#● X label name = Month Number
#● Y label name = Total profit
df_sal[['total_profit']].plot(kind='line')

plt.title('Total profit of all month')
plt.ylabel('total_profit')
plt.xlabel('month_number')

plt.show() 
```

```python
# Get total profit of all months and show line plot with the following Style properties Generated line plot must include following Style properties: –
#● Line Style dotted and Line-color should be red

df_sal[['total_units']]
print(df_sal[['total_units']].head(5))
df_sal[['total_units']].plot(kind='line', color='red', marker='o', markersize=8, markerfacecolor='red', linewidth=3, label='Total Profit')
plt.legend(loc='lower right')
plt.title('Total profit of all month')
plt.ylabel('Sold units number')
plt.xlabel('Month number')

plt.show() 
```

```python
#Read all product sales data and show it using a multiline plot
#Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product).
all_product=df_sal[["facecream","facewash","toothpaste","bathingsoap","shampoo","moisturizer"]]
print(all_product)
all_product.plot(kind='line')
plt.xlabel('Month')
plt.ylabel('Number of Units Sold')
plt.title('Number of Units Sold per Month for Each Product')
plt.show()
```

```python
#Read toothpaste sales data of each month and show it using a scatter plot. Also, add a grid in the plot. gridline style should “–“.
df_sal[['month_number','toothpaste']]
print(df_sal[['month_number','toothpaste']])

plt.scatter(df_sal['month_number'],df_sal['toothpaste'])
plt.grid(True, linestyle='--')

# Adding labels and title
plt.xlabel('Month')
plt.ylabel('Toothpaste Sales')
plt.title('Toothpaste Sales per Month')

# Display the plot
plt.show()
```

```python
# Read face cream and facewash product sales data and show it using the bar chart
df_sal[['month_number','facecream','facewash']]
print(df_sal[['month_number','facecream','facewash']])
df_sal[['facecream','facewash']].plot(kind='bar')
plt.legend(['facecream','facewash'])
plt.xlabel('month_number')
plt.ylabel('Sales')
plt.title('product sales data per Month')

# Display the plot
plt.show()
```

```python
#Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk
df_sal[['bathingsoap']]
print(df_sal[['bathingsoap']])
df_sal[['bathingsoap']].plot(kind='bar')

plt.xlabel('month_number')
plt.ylabel('Bathing soap Sales')
plt.title('Bathingsoap sales data per Month')

# Display the plot
plt.show()
```

```python
#Read the total profit of each month and show it using the histogram to see the most common profit ranges

plt.hist(df_sal['total_profit'])

# Adding labels and title
plt.xlabel('total profit')
plt.ylabel('month')
plt.title('Distribution of Total Profit per Month')

# Display the plot
plt.show()
```

```python
#Calculate total sale data for last year for each product and show it using a Pie chart
product_columns = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
total_sale_data = df_sal[product_columns].sum()
plt.pie(total_sale_data, labels=total_sale_data.index, autopct='%1.1f%%')
plt.title('Total Sale Data')
plt.show()
```

```python
#Read Bathing soap facewash of all months and display it using the Subplot
df_sal[['bathingsoap','facewash']]
print(df_sal[['bathingsoap','facewash']])
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

# Plotting Bathing Soap sales
axes[0].plot(df_sal['month_number'], df_sal['bathingsoap'], marker='o', color='blue')
axes[0].set_title('Bathing Soap Sales')
axes[0].set_xlabel('Month')
axes[0].set_ylabel('Sales')

# Plotting Facewash sales
axes[1].plot(df_sal['month_number'], df_sal['facewash'], marker='o', color='green')
axes[1].set_title('Facewash Sales')
axes[1].set_xlabel('Month')
axes[1].set_ylabel('Sales')

# Adjust layout to prevent clipping of titles
plt.tight_layout()

# Display the subplots
plt.show()
```

```python
#Read all product sales data and show it using the stack plot
plt.stackplot(df_sal['month_number'], df_sal['facecream'], df_sal['facewash'], df_sal['toothpaste'], df_sal['bathingsoap'], df_sal['shampoo'], df_sal['moisturizer'], labels=['Face Cream', 'Facewash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer'])
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.legend(loc='upper left')
plt.title('Stack Plot for All Product Sales Data')
plt.show()
```
